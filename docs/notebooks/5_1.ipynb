1. a)  Plot level lines for the function
$$
f(x) = (x_1 − 8)^2 + (x_2 - 6)^2 = r^2
$$

for different values of $r$ along with the line
$$
g(x) = x_1 + x_2 − 9 = 0
$$

using the following program. From the graph, find 

$$
\begin{split}
& f(x) \to \min\limits_{x \in \mathbb{R}^n }\\
\text{s.t. } & g(x) = 0
\end{split}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
#Plotting the circle
x=8*np.ones(8)
y=6*np.ones(8)
r=np.arange(8)/np.sqrt(2)
phi=np.linspace(0.0,2*np.pi,100)
na=np.newaxis
xline=x[na,:]+r[na,:]*np.sin(phi[:,na])
yline=y[na,:]+r[na,:]*np.cos(phi[:,na])
ax=plt.plot(xline,yline,'-')
#Plotting the line
x1=np.linspace(0,10,100)
x2=9*np.ones(100)-x1
bx=plt.plot(x1,x2,label="$x1+x2-9=0$")
plt.axis('equal')
plt.grid()
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.legend([ax[5],bx[0]],['$(x1-8)ˆ2+(x2-6)ˆ2=\\frac{25}{2}$','$x1+x2-9=0$'],loc='best')
plt.show()
```

1. c) What if we'll change the problem to:

$$
\begin{split}
& (x_1 − 8)^2 + (x_2 - 6)^2  \to \min\limits_{x \in \mathbb{R}^n }\\
\text{s.t. } & x_1 + x_2 − 9 \geq 0
\end{split}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
#Plotting the circle
x=8*np.ones(8)
y=6*np.ones(8)
r=np.arange(8)/np.sqrt(2)
phi=np.linspace(0.0,2*np.pi,100)
na=np.newaxis
xline=x[na,:]+r[na,:]*np.sin(phi[:,na])
yline=y[na,:]+r[na,:]*np.cos(phi[:,na])
ax=plt.plot(xline,yline,'-')
#Plotting the line
x1=np.linspace(-5,10,100)
x2=9*np.ones(100)-x1
x1extend=np.linspace(0,20,100)
x2extend=22*np.ones(100)-x1extend
bx=plt.plot(x1,x2,label="$x1+x2-9=0$")
plt.fill_between(x1extend,x2extend,color='grey')
plt.fill_between(x1,x2,color='white')
plt.axis('equal')
plt.grid()
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.legend(loc='best')
plt.xlim(-5,15)
plt.ylim(0,22)
#plt.savefig('../figs/2.4.eps')
plt.show()
```
